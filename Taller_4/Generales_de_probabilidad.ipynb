```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_product(rango):
    product = 1
    
    for i in range(rango):
        product *= (365-i)/365
    return product

n = []
list = []

for i in range(80):
    list.append(func_product(i))
    n.append(i)
    
plt.plot(n,list)

plt.xlabel("Personas")
plt.ylabel("P(n)")
plt.show()
```

```python
N = 100

x_1 = np.linspace(0.1, 0.9, N)
x_2 = np.linspace(0.1, 0.5, N)
respuesta = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        respuesta[i, j] = 6 * (x_1[i]) ** 2 * (1 - x_1[i]) ** 2


p_min = np.unravel_index(np.argmin(respuesta), respuesta.shape)
v_min = round(np.min(respuesta), 5)


p_max = np.unravel_index(np.argmax(respuesta), respuesta.shape)
v_max = round(np.max(respuesta), 5)

print("El valor de la probabilidad mínima es:", v_min)
print("El punto donde la probabilidad es mínima es:", p_min)
print("El valor de la probabilidad máxima es:", v_max)
print("El punto donde la probabilidad es máxima es:", p_max)


X, Y = np.meshgrid(x_1, x_2)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, respuesta, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Probabilidad')
ax.set_title('Superficie de Probabilidad')
plt.show()
```
