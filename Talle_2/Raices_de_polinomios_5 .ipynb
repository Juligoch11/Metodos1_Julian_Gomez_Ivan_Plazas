```python
import numpy as np
import sympy as sym
```

```python
def Laguerre(n):
    x = sym.symbols('x')
    f= (x**n)* sym.exp(-x)
    L=((sym.exp(x))/(np.math.factorial(n)))*f.diff(x, n)
    L=sym.expand(L)
    return L
```

```python

for i in range(21):
    poly= sym.solve(Laguerre(i))
    xi = [rt.evalf(10) for rt in poly]
    print(xi)
```
