```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n = 500
x = np.linspace(-1, 1, n + 1)
y = np.linspace(-1, 1, n + 1)
X, Y = np.meshgrid(x, y)


def semiesfera(x, y, R=1):
    total = R - (x**2 + y**2)
    return np.sqrt(np.maximum(total, 0))


esfera = semiesfera(X, Y)


fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, esfera, cmap='plasma')
plt.show()
```

```python
f,h=np.linspace(-1, 1, n + 1, retstep = True )
v = 0.0
for i in range(len(esfera)-1):
    for j in range(len(esfera)-1):
        a = 0.25 * (esfera[i, j] + esfera[i+1, j] + esfera[i, j+1] + esfera[i+1, j+1])
        ar = h**2
        v += ar * a


print(v)
```
